```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r"D:\dataset\Wine.csv")
df.head()
```

```python
from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
```

```python
x = df.iloc[:,:-1]
x
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
tsne = TSNE(n_components=2)
score = tsne.fit_transform(x)
score
```

```python
cluster.tsneplot(score, colorlist=df['Customer_Segment'], figname="Wine_t-SNE")
```
